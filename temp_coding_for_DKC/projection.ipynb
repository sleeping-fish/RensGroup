```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataDict = {'10mv-easy reduction.txt':-0.326, '20mv-easy reduction.txt':-0.338, '50mv-easy reduction.txt':-0.354, '100mv-easy reduction.txt':-0.3735, '150mv-easy reduction.txt':-0.401}
# dataDict = {'20mv.txt':-0.338}
```

```python
for file,v in dataDict.items():
    set = []
    df = pd.read_csv(r'D:\pythonProject\Python_code\temp_coding_for_DKC\easy-reduction\\'+file, delimiter='\t',header=None)
    for i in range(256):
        temp = df[df.iloc[0,0] > df.iloc[:, i] ]
        temp = temp[temp.iloc[:, i] > v]
        set.append(temp.index.size)
    dataset = np.concatenate((np.arange(len(set)).reshape(-1, 1), np.array(set).reshape(-1, 1)), axis=1)
    np.savetxt(file[:-4]+'_x.txt', dataset)
    plt.figure()
    plt.subplot(121)
    plt.title(file[:-4]+'_x')
    plt.plot(dataset[:,0], dataset[:,1])
    
    temp2 = temp.T
    df2 = df.T
    set2 = []
    for i in range(256):
        temp = df2[df2.iloc[0,0] > df2.iloc[:, i] ]
        temp = temp[temp.iloc[:, i] > v]
        set2.append(temp.index.size)
    dataset = np.concatenate((np.arange(len(set2)).reshape(-1, 1), np.array(set2).reshape(-1, 1)), axis=1)
    np.savetxt(file[:-4]+'_y.txt', dataset)
    plt.subplot(122)
    plt.title(file[:-4]+'_y')
    plt.plot(dataset[:,0], dataset[:,1])
   
```
