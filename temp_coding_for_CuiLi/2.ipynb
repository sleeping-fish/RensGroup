```python
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
filepath = r'D:\pythonProject\Data\CL\20201114 pathogen classification'
germs = ['Chryseo', 'E25922', 'EF29212', 'PA27853', 'SA29213', 'Salmonella']
```

```python
for j in range(6):
    for _, _, files in os.walk(filepath+'\\'+germs[j]):
        dataset = np.zeros((1080 * 6, 526 + 1))
        i = 0
        for file in files:
            temp = np.loadtxt(filepath+'\\'+germs[j]+'\\'+file)
            temp = temp[:1080,1].reshape(-1, 1)
            std_scaler = MinMaxScaler()
            temp = std_scaler.fit_transform(temp)
            dataset[i*1080:(i+1)*1080, :-1] = temp.flatten()
            dataset[i*1080:(i+1)*1080, -1] = i
            i += 1
```
